```python
import matplotlib.pyplot as plt

y_axis = [4010,7190, 15700, 24100, 41100, 38000, 44900, 18600]
x_axis = ["0-1990", "1991-1995", "1996-2000", "2001-2005", "2006-2010", "2011-2015", "2016-2020", "2021-2023"]

plt.bar(x_axis, y_axis)
plt.title('Articles amount on Google scholar')
plt.xlabel('year interval')
plt.ylabel('number of articles')
plt.xticks(rotation = 45) 
plt.show()

print("until 2000", 4010+7190+ 15700)
print("2001-2023", 24100+ 41100+ 38000+ 44900 + 18600)
```
